Por último de cara al modelo y para tener mejor cuadradas las estadísticas, crearemos una nueva columna para añadir las temporadas a los dos dataframes y así poder tener el dato de cada equipo y jugadores en el tiempo en el que realmente han jugado.

```python
Seasons = pd.read_csv('nbaSeasons2000-2025.csv')
```

```python
Teams = pd.read_csv('TeamStatisticsCompleted.csv')
players = pd.read_csv('PlayerStatisticsCompleted.csv')
```

Eliminamos las columnas sobrantes del ultimo merge

```python
Teams.info()
```

```python
Teams.info()
```

```python
# Convertir la columna a tipo datetime
Teams['gameDate'] = pd.to_datetime(pd.to_datetime(Teams['gameDate']).dt.date)

# Mostrar resultado
Teams.info()
```

```python
# Convertir todas las columnas de Seasons a tipo datetime
date_cols = ['Inicio Temporada Regular', 'Fin Temporada Regular', 'Inicio Playoffs', 'Fin Playoffs']
for col in date_cols:
    Seasons[col] = pd.to_datetime(Seasons[col], dayfirst=True)
```

```python
Seasons.info()
```

```python
# Función para asignar temporada
def asignar_temporada(fecha):
    for _, row in Seasons.iterrows():
        if row["Inicio Temporada Regular"] <= fecha <= row["Fin Playoffs"]:  # Se incluyen los límites correctamente
            return row["Temporada"]
    return None  # Si no hay coincidencia

# Aplicamos la función a cada fila de df2
Teams["Temporada"] = Teams["gameDate"].apply(asignar_temporada)

# Mostrar filas donde no se asignó temporada (para verificar)
df_sin_temporada = Teams[Teams["Temporada"].isna()]
print(df_sin_temporada)

# Si aún hay valores None, revisamos el mínimo y máximo de gameDate
print("Fecha más antigua en df2:", Teams["gameDate"].min())
print("Fecha más reciente en df2:", Teams["gameDate"].max())

# Verificar las fechas mínimas y máximas en df1 para asegurarnos de que cubren el rango
print("Fecha más antigua en df1:", Seasons["Inicio Temporada Regular"].min())
print("Fecha más reciente en df1:", Seasons["Fin Playoffs"].max())
```

Como los salarios los tenemos a partir de la temporada 2000-2001, eliminamos los registros correspondientes a la temporada 1999-2000

```python
# Encuentra los índices de las filas que contienen la temporada 1999-2000
indices_a_eliminar = Teams[Teams['Temporada'] == '1999-2000'].index

# Elimina las filas correspondientes
Teams = Teams.drop(indices_a_eliminar)

# Imprime el DataFrame resultante (opcional)
Teams
```
